```python
list = [3,7,1,2,2,4]
def sum_prime(list):
    sum = 0
    for i in list:
        if i>3 or i<=1:
            for j in range(2,int(i**0.5)+1):
                if (i%j == 0):
                    break
                else:
                    sum+=i
                    break
        else:
            sum+=i                
    return sum
sum_prime(list)
```

```python
list = [5,6,7,9,11,10]
def increasing(list):
    for i in range(0,len(list)-1):
        if list[i]<list[i+1]:
            continue
        else:
            return False
    return True
increasing(list)
```

```python
list=[3,4,15,120,223]
def expanding(list):
    d = list[1]-list[0]
    for i in range(1,len(list)-1):
            if (list[i+1]-list[i]>d):
                continue
            else:
                return False
    return True
expanding(list)
```
